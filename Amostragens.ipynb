# **Amostragens**

### Importando as Bibliotecas


```python
import pandas as pd
import numpy as np

import warnings


warnings.filterwarnings('ignore')

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/enem_2019_tratados.csv', sep=',', encoding='iso-8859-1')
df
```

```python
enem_campinas = df.loc[df.NO_MUNICIPIO_RESIDENCIA == 'Campinas']
enem_campinas
```

### **Amostragem aleatórias simples**

```python
np.random.seed(1)
```

```python
amos_simples = enem_campinas.sample(frac=0.20) # Porcentagem
amos_simples
```

```python
amos_simples = enem_campinas.sample(3000) # Quantidade numérica
amos_simples
```
